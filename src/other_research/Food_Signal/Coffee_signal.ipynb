```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Coffee Datasetï¼š

```python
def coffee_data():
    
    # Load a TXT file and transform it into a DataFrame.
    
    data_train = pd.read_csv('../../../UCR_datasets/Coffee_TRAIN.txt', sep = '\s+', header = None)
    data_test = pd.read_csv('../../../UCR_datasets/Coffee_TEST.txt', sep = '\s+', header = None)
    
    # Separate the target variable from the time series, convert the target variable to integers, and convert the time series data to floats.
    
    train_targets = data_train.iloc[:, 0].astype(int)  
    train_signals = data_train.iloc[:, 1:].astype(float)  

    test_targets = data_test.iloc[:, 0].astype(int)  
    test_signals = data_test.iloc[:, 1:].astype(float)  

    # Transform the time series data of each row into a list representation.
    
    train_signalss = []
    test_signalss = []
    
    train_signals_list = train_signals.values.tolist()
    
    for row in train_signals_list:
        train_signalss.append(row)
    
    test_signals_list = test_signals.values.tolist()
    
    for row in test_signals_list:
        test_signalss.append(row)

    # Construct a new DataFrame with the 'signal' column represented as lists.
    
    df_train = pd.DataFrame({
        'signal': train_signalss,
        'target': train_targets
    })

    df_test = pd.DataFrame({
        'signal': test_signalss,
        'target': test_targets
    })

    return df_train, df_test
```

```python
df_train, df_test = coffee_data()
```

```python
# Class 1 ###########################################

signal_1 = []

train1 = df_train['signal'][0:14].values.tolist()

test1 = df_test['signal'][0:15].values.tolist()

for row in train1:
    signal_1.append(row)

for row in test1:
    signal_1.append(row)

signal1 = np.mean(signal_1, axis = 0)

# Class 2 ###########################################

signal_2 = []

train2 = df_train['signal'][14:28].values.tolist()

test2 = df_test['signal'][15:28].values.tolist()

for row in train2:
    signal_2.append(row)

for row in test2:
    signal_2.append(row)

signal2 = np.mean(signal_2, axis=0)
```

# signal Diagram

```python
plt.figure(figsize = (24, 8), dpi = 1000)

plt.subplot(1, 2, 1)

plt.plot(signal1, color = 'midnightblue', linestyle = '--', marker = 's', markerfacecolor = 'none', label = 'class 1', markevery = 10)

plt.plot(signal2, color = 'deeppink', linestyle = '-.', marker = 'o', markerfacecolor = 'none', label = 'class 2', markevery = 10)

plt.xlabel('Time series index', fontsize = 25, labelpad = 20, fontweight = 'bold')
plt.ylabel('Absorbance', fontsize = 25, labelpad = 15, fontweight = 'bold')

plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)

plt.legend(fontsize = 15)

plt.subplot(1, 2, 2)

signal2 = signal2 - 3

plt.plot(signal1, color = 'midnightblue', linestyle = '--', marker = 's', markerfacecolor = 'none', label = 'class 1', markevery = 10)
plt.plot(signal2, color = 'deeppink', linestyle = '-.', marker = 'o', markerfacecolor = 'none', label = 'class 2', markevery = 10)

plt.xticks(fontsize = 20)
plt.yticks([])

plt.xlabel('Time series index', fontsize = 25, labelpad = 20, fontweight = 'bold')

plt.legend(fontsize = 15)

plt.subplots_adjust(left = 0.07, right = 0.98, wspace = 0.06) 

plt.savefig('../../../images/coffee.png', bbox_inches = 'tight', dpi = 1000)
# plt.savefig('pics/coffee.svg', bbox_inches='tight', dpi = 1000)

plt.show()
```
