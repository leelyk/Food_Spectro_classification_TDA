```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## strawberry Dataset：

```python
def strawberry_data():
    
    # Load a TXT file and transform it into a DataFrame.
    
    data_train = pd.read_csv('../../../UCR_datasets/Strawberry_TRAIN.txt', sep = '\s+', header = None)
    data_test = pd.read_csv('../../../UCR_datasets/Strawberry_TEST.txt', sep = '\s+', header = None)
    
    # Separate the target variable from the time series, convert the target variable to integers, and convert the time series data to floats.
    
    train_targets = data_train.iloc[:, 0].astype(int)  
    train_signals = data_train.iloc[:, 1:].astype(float)  

    test_targets = data_test.iloc[:, 0].astype(int)  
    test_signals = data_test.iloc[:, 1:].astype(float)  

    # Transform the time series data of each row into a list representation.
    
    train_signalss = []
    test_signalss = []
    
    train_signals_list = train_signals.values.tolist()
    
    for row in train_signals_list:
        train_signalss.append(row)
    
    test_signals_list = test_signals.values.tolist()
    
    for row in test_signals_list:
        test_signalss.append(row)

    # Construct a new DataFrame with the 'signal' column represented as lists.
    
    df_train = pd.DataFrame({
        'signal': train_signalss,
        'target': train_targets
    })

    df_test = pd.DataFrame({
        'signal': test_signalss,
        'target': test_targets
    })

    return df_train, df_test
```

```python
df_train, df_test = strawberry_data()
```

```python
# Class 1 ###########################################

signal_1 = []

train1 = df_train['signal'][64:72].values.tolist()

for row in (df_train['signal'][80:118].values.tolist()):
    train1.append(row)
for row in (df_train['signal'][164:210].values.tolist()):
    train1.append(row)
for row in (df_train['signal'][230:240].values.tolist()):
    train1.append(row)
for row in (df_train['signal'][382:407].values.tolist()):
    train1.append(row)
for row in (df_train['signal'][415:452].values.tolist()):
    train1.append(row)
for row in (df_train['signal'][497:543].values.tolist()):
    train1.append(row)
for row in (df_train['signal'][561:570].values.tolist()):
    train1.append(row)

test1 = df_test['signal'][0:24].values.tolist()

for row in (df_test['signal'][31:68].values.tolist()):
    test1.append(row)
for row in (df_test['signal'][112:157].values.tolist()):
    test1.append(row)
for row in (df_test['signal'][173:182].values.tolist()):
    test1.append(row)
for row in (df_test['signal'][353:370].values.tolist()):
    test1.append(row)

for row in train1:
    signal_1.append(row)

for row in test1:
    signal_1.append(row)

signal1 = np.mean(signal_1, axis = 0)

# Class 2 ###########################################

signal_2 = []

train2 = df_train['signal'][0:64].values.tolist()

for row in (df_train['signal'][72:80].values.tolist()):
    train2.append(row)
for row in (df_train['signal'][118:164].values.tolist()):
    train2.append(row)
for row in (df_train['signal'][210:230].values.tolist()):
    train2.append(row)
for row in (df_train['signal'][240:382].values.tolist()):
    train2.append(row)
for row in (df_train['signal'][407:415].values.tolist()):
    train2.append(row)
for row in (df_train['signal'][452:497].values.tolist()):
    train2.append(row)
for row in (df_train['signal'][543:561].values.tolist()):
    train2.append(row)
for row in (df_train['signal'][570:613].values.tolist()):
    train2.append(row)

test2 = df_test['signal'][51:105].values.tolist()

for row in (df_test['signal'][24:31].values.tolist()):
    test2.append(row)
for row in (df_test['signal'][68:112].values.tolist()):
    test2.append(row)
for row in (df_test['signal'][157:173].values.tolist()):
    test2.append(row)
for row in (df_test['signal'][182:353].values.tolist()):
    test2.append(row)

for row in train2:
    signal_2.append(row)

for row in test2:
    signal_2.append(row)

signal2 = np.mean(signal_2, axis = 0)
```
