```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Beef Dataset：

```python
def beef_data():
    
    # Load a TXT file and transform it into a DataFrame.
    
    data_train = pd.read_csv('../../../UCR_datasets/Beef_TRAIN.txt', sep = '\s+', header = None)
    data_test = pd.read_csv('../../../UCR_datasets/Beef_TEST.txt', sep = '\s+', header = None)
    
    # Separate the target variable from the time series, convert the target variable to integers, and convert the time series data to floats.
    
    train_targets = data_train.iloc[:, 0].astype(int)  
    train_signals = data_train.iloc[:, 1:].astype(float)  

    test_targets = data_test.iloc[:, 0].astype(int)  
    test_signals = data_test.iloc[:, 1:].astype(float)  

    # Transform the time series data of each row into a list representation.
    
    train_signalss = []
    test_signalss = []
    
    train_signals_list = train_signals.values.tolist()
    
    for row in train_signals_list:
        train_signalss.append(row)
    
    test_signals_list = test_signals.values.tolist()
    
    for row in test_signals_list:
        test_signalss.append(row)

    # Construct a new DataFrame with the 'signal' column represented as lists.
    
    df_train = pd.DataFrame({
        'signal': train_signalss,
        'target': train_targets
    })

    df_test = pd.DataFrame({
        'signal': test_signalss,
        'target': test_targets
    })

    return df_train, df_test
```

```python
df_train, df_test = beef_data()
```

```python
# Class 1 ###########################################

signal_1 = []

train1 = df_train['signal'][0:6].values.tolist()

test1 = df_test['signal'][0:6].values.tolist()

for row in train1:
    signal_1.append(row)

for row in test1:
    signal_1.append(row)

signal1 = np.mean(signal_1, axis = 0)

# Class 2 ###########################################

signal_2 = []

train2 = df_train['signal'][6:12].values.tolist()

test2 = df_test['signal'][6:12].values.tolist()

for row in train2:
    signal_2.append(row)

for row in test2:
    signal_2.append(row)

signal2 = np.mean(signal_2, axis = 0)

# Class 3 ###########################################

signal_3 = []

train3 = df_train['signal'][12:18].values.tolist()

test3 = df_test['signal'][12:18].values.tolist()

for row in train3:
    signal_3.append(row)

for row in test3:
    signal_3.append(row)

signal3 = np.mean(signal_3, axis=0)

# Class 4 ###########################################

signal_4 = []

train4 = df_train['signal'][18:24].values.tolist()

test4 = df_test['signal'][18:24].values.tolist()

for row in train4:
    signal_4.append(row)

for row in test4:
    signal_4.append(row)

signal4 = np.mean(signal_4, axis=0)

# Class 5 ###########################################

signal_5 = []

train5 = df_train['signal'][24:30].values.tolist()

test5 = df_test['signal'][24:30].values.tolist()

for row in train5:
    signal_5.append(row)

for row in test5:
    signal_5.append(row)

signal5 = np.mean(signal_5, axis=0)
```

# signal Diagram

```python
plt.figure(figsize = (24, 8), dpi = 1000)

plt.subplot(1, 2, 1)

plt.plot(signal1, color = 'midnightblue', linestyle = '--', marker = '*', label = 'class 1', markevery = 30)

plt.plot(signal2, color = 'deeppink', linestyle = '-.', marker = 'o', markerfacecolor = 'none', label = 'class 2', markevery = 30)

plt.plot(signal3, color = 'darkorange', linestyle = ':', marker = 'D', markerfacecolor = 'none', label = 'class 3', markevery = 30)

plt.plot(signal4, color = 'teal', linestyle = '-', marker = 'P', markerfacecolor = 'none', label = 'class 4', markevery = 30)

plt.plot(signal5, color = 'purple', linestyle = ':', marker = 'X', markerfacecolor = 'none', label = 'class 5', markevery = 30)

plt.xlabel('Time series index', fontsize = 25, labelpad = 20, fontweight = 'bold')
plt.ylabel('Absorbance', fontsize = 25, labelpad = 15, fontweight = 'bold')

plt.xticks([0, 100, 200, 300, 400, 500, 600], fontsize = 20)

plt.yticks(fontsize = 20)

plt.legend(fontsize = 15)

plt.subplot(1, 2, 2)

signal1 = signal1 + 10
signal2 = signal2 + 5
signal4 = signal4 - 5
signal5 = signal5 - 10

plt.plot(signal1, color = 'midnightblue', linestyle = '--', marker = '*', label = 'class 1', markevery = 30)

plt.plot(signal2, color = 'deeppink', linestyle = '-.', marker = 'o', markerfacecolor = 'none', label = 'class 2', markevery=30)

plt.plot(signal3, color = 'darkorange', linestyle = ':', marker = 'D', markerfacecolor = 'none', label = 'class 3', markevery=30)

plt.plot(signal4, color = 'teal', linestyle = '-', marker = 'P', markerfacecolor = 'none', label = 'class 4', markevery=30)

plt.plot(signal5, color = 'purple', linestyle = ':', marker = 'X', markerfacecolor = 'none', label = 'class 5', markevery=30)

plt.yticks([])

plt.xlabel('Time series index', fontsize = 25, labelpad = 20, fontweight = 'bold')

plt.xticks([0, 100, 200, 300, 400, 500, 600], fontsize = 20)

plt.legend(fontsize = 15)

plt.subplots_adjust(left = 0.07, right = 0.98, wspace = 0.06) # 增加間距

plt.savefig('../../../images/beef.png', bbox_inches = 'tight', dpi = 1000)
# plt.savefig('pics/beef.svg', bbox_inches='tight', dpi = 1000)

plt.show()
```
